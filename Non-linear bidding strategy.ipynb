# The Optimal Real-Time bidding strategy

```python
# ORTB Bidding Function on validation set:
# ORTB = *√ c / λ x θ + c^2 − c
c_scope = np.arange(1,16,1)
lam_scope = [1e-10,5e-10,1e-9,5e-9,1e-8,5e-8,1e-7,5.2e-7,1e-6,5e-6,
             1e-5,5e-5,1e-4,5e-4,1e-3,5e-3,1e-2,5e-2,1e-1,5e-1]

ortb_bids = []
parameters = []

# seek for the all combinations of parameter to get the best ortbbids
for c in c_scope:
    for lam in lam_scope:
        parameters.append((c, lam))
        inner_eq = ((c/lam * p_CTR) + c**2 - c) 
        ortb_bid = np.sqrt(inner_eq)
        ortb_bids.append(ortb_bid.values.tolist())
```

```python
print("The length of the parameters is:", len(parameters)) #20*15
print("The length of the ortb_bids is:", len(ortb_bids))
```

```python
import time
def ortb_bidding(bid_won):
    clicks = 0
    impressions = 0.0
    budget = 6250 * 1000
    cost = 0.0

    # bidding win procedure
    bid_won >= df_val.payprice
    for i in range(0, len(bid_won)):
        if bid_won[i] == True:
            clicks = clicks + df_val.click[i]
            impressions = impressions + 1.0
            cost = df_val.payprice[i]
            print("the bid is Within-Budget")
        if cost >= budget:
            print("the bid is Over-Budget")
            break
    return impressions, clicks, cost
```

```python
ORTB = pd.DataFrame()
ORTB['C, Lambda'] = parameters

Impressions = []
Clicks = []
Cost = []

starttime = time.time()
print("Start ortb bidding...")

for n in ortb_bids:
    [imps, clks, ct] = ortb_bidding(n)
    Impressions.append(imps)
    Clicks.append(clks)
    Cost.append(ct)

ORTB['impressions'] = Impressions
# impression can't be a float
ORTB.impressions = ORTB.impressions.astype(int)
ORTB['cost'] = Cost
ORTB['clicks'] = Clicks
ORTB['CTP'] = (ORTB.clicks / ORTB.impressions * 100).round(4).astype(str)
ORTB['CPM'] = (ORTB.cost / ORTB.impressions * 1000).round(2).astype(str)
ORTB['CPC'] = (ORTB.cost / ORTB.clicks).round(2).astype(str)

print("ORTB bidding takes %s minutes" % round(((time.time() - starttime)/60),2) + "to finish")
```
